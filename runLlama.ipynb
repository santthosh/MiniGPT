### Step 2: Load LLaMA Model and Tokenizer

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Choose a LLaMA model (change the model name if needed)
model_name = "meta-llama/Llama-2-7b-chat-hf"  # Change this to another LLaMA model if needed

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model (with fp16 for efficiency if using CUDA)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16 if device=="cuda" else torch.float32)
model.to(device)

print("LLaMA model loaded successfully!")
```

### Step 3: Generate Text

```python
def generate_text(prompt, max_length=100):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=max_length,
            temperature=0.7,  # Adjust randomness
            top_k=50,  # Limit to top 50 choices
            top_p=0.9,  # Use nucleus sampling
            do_sample=True  # Enable sampling
        )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example usage
prompt = "Hello world"
generated_text = generate_text(prompt)
print("Generated Text:\n", generated_text)
```
